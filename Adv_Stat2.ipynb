Ques 1. What is hypothesis testing in statistics?

Ans. In hypothesis testing, we start with a null hypothesis (H₀), which represents a statement of no effect or no difference, and an alternative hypothesis (H₁ or Ha), which represents a statement of effect or difference. The goal is to use sample data to determine whether there is enough statistical evidence to reject the null hypothesis in favor of the alternative hypothesis.

Ques 2. What is null hypothesis, and how does it differ from alternative hypothesis?

Ans. In statistical hypothesis testing, the null hypothesis (H0) asserts that there is no significant relationship or difference between variables, suggesting that any observed effects are due to random chance. The alternative hypothesis (Ha) or (H1) is the statement that there is a significant relationship or difference between variables, and that the observed effects are not due to chance. In essence, the null hypothesis represents the status quo, while the alternative hypothesis represents the researcher's proposed explanation or finding.

Null Hypothesis (H0):

States no effect or relationship: It assumes that any observed differences or patterns are merely due to random variation in the data.
What is tested: Researchers typically try to disprove or reject the null hypothesis.
Example: "There is no difference in the average test scores between students who use method A and those who use method B".

Alternative Hypothesis (Ha or H1):

States a relationship or difference:
It proposes that there is a real effect or relationship between variables, and that the observed differences are not due to chance.
What is tested:
Researchers attempt to find evidence to support the alternative hypothesis, which is accepted if the null hypothesis is rejected.
Example:
"Students who use method A have higher average test scores than students who use method B".

Key Differences:
Purpose:
The null hypothesis is the default position, assumed to be true until evidence is presented to reject it. The alternative hypothesis is the researcher's proposed explanation or finding, which is accepted if the null hypothesis is rejected.

Relationship:
The null hypothesis and alternative hypothesis are mutually exclusive; if one is true, the other must be false.

Notation:
The null hypothesis is typically denoted as H0, while the alternative hypothesis is denoted as Ha or H1.

Testing:
Researchers use statistical tests to evaluate evidence and determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

Ques 3. What is the significance level in hypothesis testing, and why is it important?

Ans. In hypothesis testing, the significance level, denoted as α (alpha), is a threshold set by the researcher before conducting the test. It represents the probability of rejecting the null hypothesis when it is actually true, known as a Type I error. In other words, it's the risk you're willing to take for making a false positive conclusion.

Importance of the Significance Level: Controlling Type I Errors: By setting a significance level, researchers control the likelihood of making incorrect conclusions. Balancing Errors: Choosing the right significance level helps balance the risk of Type I errors against the need for sensitivity to detect actual effects. Contextual Relevance: The choice of significance level depends on the context of the study and the consequences of making errors. In medical research, a lower significance level (e.g., 0.01) is often used to minimize the risk of false positives. In exploratory research, a higher significance level (e.g., 0.10) might be acceptable.

Ques 4. What does a P-value represent in hypothesis testing?

Ans. The p-value indicates how likely it is to observe the collected data, or something more extreme, under the assumption that the null hypothesis is correct. A smaller p-value suggests stronger evidence against the null hypothesis. For example, a p-value of 0.01 means there's a 1% chance of observing the data if the null hypothesis is true.

Ques 5. How do you interpret the P-value in hypothesis testing?

Ans. In hypothesis testing, the p-value is a statistical measure that helps determine the strength of the evidence against the null hypothesis (H₀). It represents the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. Interpreting the P-Value: P-value ≤ α (e.g., 0.05): Reject the null hypothesis; the result is statistically significant. P-value > α: Fail to reject the null hypothesis; the result is not statistically significant. Here, α (alpha) is the pre-determined significance level, commonly set at 0.05, representing a 5% risk of concluding that a difference exists when there is none.

Ques 6. What are Type1 and Type2 errors in hypothesis testing?

Ans. Type I Error (False Positive): Definition: Occurs when the null hypothesis (H₀) is true, but is incorrectly rejected. Consequence: Concluding that there is an effect or difference when, in reality, there isn't. Probability: Denoted by α (alpha), often set at 0.05, indicating a 5% risk of making this error. Example: A clinical trial concludes that a new drug is effective when it actually isn't.

Type II Error (False Negative): Definition: Occurs when the null hypothesis is false, but is incorrectly not rejected. Consequence: Failing to detect an effect or difference that actually exists. Probability: Denoted by β (beta). The power of a test is calculated as 1 - β, representing the probability of correctly rejecting a false null hypothesis. Example: A clinical trial fails to detect the effectiveness of a new drug that actually works.

Ans. One-Tailed Test: Definition: Tests for the possibility of an effect in one direction (either greater than or less than a certain value).

Investopedia Critical Region: Located entirely in one tail of the probability distribution Significance Level (α): All of α is allocated to one tail. It's a method for hypothesis testing, allowing researchers to decide if observed differences are likely due to chance or a real effect.

When to use a z-test:

Known Population Standard Deviation:
You must know the population standard deviation (sigma) or have a large enough sample size (n > 30) to effectively use the sample standard deviation as an estimate.

Large Sample Size:
Generally, the sample size should be greater than 30 to ensure the test statistic follows a normal distribution.

Normally Distributed Population (or Large Sample Size):
The population from which the sample is drawn should be normally distributed, or if the sample size is large, the Central Limit Theorem can help ensure the sampling distribution of the mean is approximately normal.

Comparing to a Known Value:
One-sample z-tests are used to compare a sample mean to a known population mean.

Comparing Two Groups:
Two-sample z-tests are used to compare the means of two different populations or groups.


Ques 9. How do you calculate the Z-score,and what does it represent in hypothesis testing?

Ans. A Z-score represents how many standard deviations a particular data point is from the mean of a dataset. It's calculated by subtracting the mean from the data point and then dividing by the standard deviation. In hypothesis testing, a Z-score is used to determine the probability of observing a data point (or a sample mean) if the null hypothesis were true.

Calculating the Z-score:

Identify the data point (x) you want to standardize.
Determine the mean (μ) and standard deviation (σ) of the dataset. If you're using a sample, you'll use the sample mean (x̄) and sample standard deviation (s).
Apply the Z-score formula:
z = (x - μ) / σ: (when using population parameters)
z = (x - x̄) / s: (when using sample parameters)

What a Z-score Represents in Hypothesis Testing:

Distance from the Mean:
A Z-score indicates how many standard deviations away from the mean a data point (or sample mean) is.

Probability:
A Z-score is used to calculate a p-value, which represents the probability of observing the data (or a more extreme result) if the null hypothesis is true. A small p-value (usually less than the significance level, α) suggests strong evidence against the null hypothesis.

Decision Rule:
In hypothesis testing, the Z-score is compared to a critical value determined by the significance level. If the Z-score falls into the rejection region (beyond the critical value), the null hypothesis is rejected.

Example:
Imagine you are testing whether a new drug improves the blood pressure of patients. You collect data on a group of patients and calculate the Z-score for the sample mean blood pressure. A larger Z-score (further from the mean of the population) would suggest that the new drug has a significant effect on blood pressure, and you would be more likely to reject the null hypothesis that the drug has no effect.

Ques 12. What is the T-test,and how is it used in hypothesis testing?

Ans. A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups. It's particularly useful when comparing the means of two samples, or comparing a sample mean to a known population mean, especially when the population standard deviation is unknown. In hypothesis testing, the t-test helps determine if observed differences between sample means are likely due to a true difference in the population or due to random chance.
How it's used in hypothesis testing:
1. Formulate Hypotheses:
Null Hypothesis (H0): There is no significant difference between the means of the two groups or between the sample mean and the population mean.
Alternative Hypothesis (H1): There is a significant difference between the means.
2. Set Significance Level (α):
This is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01.
3. Calculate the Test Statistic:
The t-test calculates a t-value, which measures the difference between the sample means (or the sample mean and population mean) relative to the variability within the samples.
4. Determine Degrees of Freedom:
Degrees of freedom (df) depend on the type of t-test (e.g., independent samples t-test, paired t-test) and the sample sizes.
5. Find the Critical Value(s):
Using the t-distribution table and the significance level (α), determine the critical t-value(s) for the specified degrees of freedom.
6. Make a Decision:
If the calculated t-value is greater than the critical value(s), reject the null hypothesis. This indicates that there's a statistically significant difference between the means.
If the calculated t-value is not greater than the critical value(s), fail to reject the null hypothesis. This suggests that the observed difference may be due to random chance.

Ques 14. What is a confidence interval,and how is it used to interpret statistical results?

Ans. A confidence interval is a range of values within which the true population parameter is likely to fall, based on a sample. It's a statistical tool used to estimate the reliability of a sample statistic and quantify the uncertainty surrounding the estimate. A confidence level, typically 95% or 99%, indicates the probability that the true population parameter will fall within the interval if the process of constructing confidence intervals were repeated many times.

How to interpret statistical results using confidence intervals:

1. Understanding the interval:
A 95% confidence interval means that if you repeated your sampling process 100 times, you would expect that 95 out of those 100 intervals would contain the true population parameter. It doesn't mean there's a 95% chance that the true value is within the interval for a single, specific interval.

2. Comparing to a null hypothesis:
If the confidence interval includes the hypothesized value (the null hypothesis), you cannot reject the null hypothesis with the given confidence level. For example, if you are testing the null hypothesis that a new treatment has no effect on blood pressure, and the confidence interval for the difference in blood pressure includes zero, you wouldn't reject the null hypothesis.

3. Interpreting effect size:
Confidence intervals can also be used to estimate the magnitude of an effect. For instance, if you're studying the effect of an intervention, the confidence interval around the difference in the outcome group and control group can give you a sense of how large the effect is likely to be.

4. Using confidence intervals in hypothesis testing:
Confidence intervals can be used as an alternative to p-values in hypothesis testing. If a confidence interval does not include the null hypothesis value, it suggests that the result is statistically significant.

Ques 15. What is the margin of error,and how does it affect the confidence interval?

Ans. The margin of error is a statistical measure that reflects the precision of an estimate, essentially how much the sample results might differ from the real population results. It's half the width of a confidence interval. A smaller margin of error indicates a more precise estimate, meaning the sample results are likely closer to the true population value. The confidence interval, which represents the range within which the true population value is likely to fall, is directly affected by the margin of error.

1. What is Margin of Error?
The margin of error is expressed as a plus or minus percentage (e.g., +/- 3 percentage points).
It tells you how much the survey results might differ from the true population value.
A smaller margin of error indicates a more precise estimate, meaning the survey results are likely closer to the real population values.

2. How does Margin of Error Affect the Confidence Interval?
The confidence interval is the range within which the true population value is likely to fall, with a specified level of confidence (e.g., 95%).
The margin of error is one of the components of the confidence interval calculation.

A larger margin of error means a wider confidence interval. This is because the larger the margin, the wider the range you need to be confident that the true population value is within.
Conversely, a smaller margin of error means a narrower confidence interval. You can be more confident that the true population value falls within a smaller range when the margin of error is smaller.
The confidence level also plays a role. A higher confidence level (e.g., 99%) requires a wider confidence interval (and thus a larger margin of error) to account for a greater degree of certainty.

Ques 18. What is the Chi-square goodness of fit test,and how is it applied?

Ans. The chi-square goodness of fit test is a statistical test used to determine how well a sample data matches a population with a specific distribution. It helps assess whether the observed data frequencies are significantly different from the expected frequencies based on a theoretical distribution.

How it works:
1. Hypothesis:
You start with a hypothesis about the population's distribution. For example, you might hypothesize that a population follows a normal distribution.

2. Data Collection:
Collect a sample from the population.

3. Expected Frequencies:
Based on your hypothesis, calculate the expected frequencies for different categories or bins of the sample data.

4. Observed Frequencies:
Determine the actual observed frequencies for each category or bin in your sample.

5. Chi-Square Statistic:
Calculate the chi-square statistic, which measures the discrepancy between the observed and expected frequencies. The formula is: χ² = Σ [(O - E)² / E], where O is the observed frequency, and E is the expected frequency.

6. Degrees of Freedom:
Calculate the degrees of freedom, which depend on the number of categories or bins.

7. P-value:
Find the p-value associated with the chi-square statistic and degrees of freedom using a chi-square distribution table or statistical software.

8. Decision:
Compare the p-value with a chosen significance level (e.g., 0.05). If the p-value is less than the significance level, you reject the null hypothesis (the population follows the hypothesized distribution), indicating a poor fit. If the p-value is greater than the significance level, you fail to reject the null hypothesis, suggesting a good fit.

Ques 20. What is an ANOVA test,and what are its assumptions?

Ans. An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of two or more groups to determine if there is a significant difference between them. It works by analyzing the variance within each group and between groups, ultimately determining if the variance between group means is greater than expected by chance. The assumptions of ANOVA include random sampling, independence of observations, normality of residuals, and homogeneity of variances.

Elaboration:
1. What is an ANOVA test?

Purpose:
ANOVA tests the null hypothesis that the means of all groups are equal. It's particularly useful when comparing more than two groups, as multiple t-tests (which compare two groups) can increase the risk of a Type I error (falsely concluding a difference when there isn't one).

Mechanism:
ANOVA analyzes the variance (spread) of the data, comparing the variance within each group (how spread out the data points are within a group) to the variance between the group means (how much the group means differ from each other).

2. Assumptions of ANOVA:

Independence of Observations: Each observation should be independent of others within and between groups. This means the value of one observation should not influence another.

Random Sampling: Observations within each group should be sampled randomly and independently.
Normality of Residuals: The residuals (the difference between observed and predicted values) should be normally distributed within each group.

Homogeneity of Variances (Homoscedasticity): The variance within each group should be approximately equal.

Interval or Ratio Level Data: The dependent variable should be measured at an interval or ratio level.


Ques 21. What are the different types of ANOVA test?

Ans. There are two main types of ANOVA tests: one-way ANOVA and two-way ANOVA. One-way ANOVA compares the means of three or more groups based on a single independent variable, while two-way ANOVA analyzes the effects of two or more independent variables on a dependent variable. Additionally, there are variations like repeated measures ANOVA and MANOVA, which are extensions of the basic ANOVA.

One-way ANOVA:
This test is used when you want to compare the means of three or more groups or levels of a single independent variable. For example, you could use a one-way ANOVA to compare the average weight loss among different diets.

Two-way ANOVA:
This test is used when you want to examine the effects of two or more independent variables on a dependent variable, including potential interactions between the independent variables. For instance, you could use a two-way ANOVA to compare the productivity of workers based on both salary and their skill set.

Repeated Measures ANOVA:
This test is used when you want to compare the means of three or more groups where the same subjects are measured multiple times under different conditions or at different time points.

MANOVA (Multivariate Analysis of Variance):
This test is used when you have multiple dependent variables.

ANCOVA (Analysis of Covariance):
This test is used when you want to control for the effects of one or more covariate variables in the analysis.

Ques 22. What is the F-test,and how does it relate to hypothesis testing?

Ans. The F-test is a statistical test used to compare variances between two or more groups. It's often part of the analysis of variance (ANOVA) and is used to determine if there's a significant difference between the means of those groups. In essence, the F-test helps determine if the differences in group means are due to random variation or a real effect.

How it relates to hypothesis testing:
Null Hypothesis:
The F-test, like other hypothesis tests, starts with a null hypothesis. In the context of ANOVA, the null hypothesis typically states that there is no difference between the population means of the groups being compared.

Alternative Hypothesis:
The alternative hypothesis suggests that there is a significant difference between the population means of at least two of the groups.

F-statistic:
The F-test calculates an F-statistic, which is a ratio of variances. This statistic is then compared to a critical value from the F-distribution to determine if the null hypothesis should be rejected.

F-distribution:
The F-test relies on the F-distribution, a probability distribution that describes the distribution of the F-statistic under the assumption that the null hypothesis is true.

Decision Rule:
If the calculated F-statistic is greater than the critical value from the F-distribution, the null hypothesis is rejected, indicating a statistically significant difference between the group means.

p-value:
The F-test can also provide a p-value, which represents the probability of observing the calculated F-statistic (or a more extreme value) if the null hypothesis is true. A low p-value (typically less than 0.05) provides evidence to reject the null hypothesis.

```python
# Ques 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
# distribution.

# Solution 14.

import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, data_mean, data_std):
  """
    Calculate the Z-score for a given data point.

    Args:
      data_point (float or int): The data point to calculate the Z-score for.
      data_mean (float or int): The mean of the dataset.
      data_std (float or int): The standard deviation of the dataset.

    Returns:
      float: The Z-score for the data point.
  """
  z_score = (data_point - data_mean) / data_std
  return z_score

def compare_to_standard_normal(z_score):
  """
    Compare the Z-score to a standard normal distribution.

    Args:
      z_score (float): The Z-score to compare.

    Returns:
      float: The probability of observing a value as extreme or more extreme than the given Z-score.
  """
  p_value = 2 * (1 - norm.cdf(abs(z_score)))
  return p_value

if __name__ == "__main__":
  data = [70, 85, 90, 60, 75, 80, 95, 88, 78, 82]
  data_point = 92
  data_mean = np.mean(data)
  data_std = np.std(data)

  z_score = calculate_z_score(data_point, data_mean, data_std)
  p_value = compare_to_standard_normal(z_score)

  print(f"Data point: {data_point}")
  print(f"Mean: {data_mean:.2f}")
  print(f"Standard deviation: {data_std:.2f}")
  print(f"Z-score: {z_score:.2f}")
  print(f"P-value: {p_value:.3f}")

  alpha = 0.05
  if p_value < alpha:
    print(f"The Z-score is statistically significant at the {alpha} level.")
  else:
    print(f"The Z-score is not statistically significant at the {alpha} level.")
```

```python
# Ques 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

# Solution 3.

import numpy as np
from scipy.stats import norm

def one_sample_ztest(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test.

    Args:
        sample (list or numpy.ndarray): The sample data.
        pop_mean (float): The population mean.
        pop_std (float): The population standard deviation.
        alpha (float, optional): Significance level. Defaults to 0.05.
        alternative (str, optional): 'two-sided', 'less', or 'greater'.
            Defaults to 'two-sided'.

    Returns:
        tuple: (z-statistic, p-value)
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    if alternative == 'two-sided':
        p_value = 2 * norm.cdf(-np.abs(z_stat))
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:
        raise ValueError("Alternative must be 'two-sided', 'less', or 'greater'")

    return z_stat, p_value

if __name__ == '__main__':
    # Example Usage
    sample_data = [78, 85, 92, 80, 88, 83, 75, 95, 87, 81]
    population_mean = 80
    population_std_dev = 8

    z_statistic, p_value = one_sample_ztest(sample_data, population_mean, population_std_dev)

    print("Z-statistic:", z_statistic)
    print("P-value:", p_value)

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
```

```python
# Ques 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

# Solution 4.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters
alpha = 0.05  # Significance level
mean_null = 0  # Mean under the null hypothesis
std_dev = 1    # Population standard deviation
sample_mean = 0.2 # Sample mean
sample_size = 100 # Sample size

# Calculate the Z-score
z_score = (sample_mean - mean_null) / (std_dev / np.sqrt(sample_size))

# Determine critical values for a two-tailed test
critical_value = norm.ppf(1 - alpha/2)

# Calculate p-value
p_value = 2 * (1 - norm.cdf(np.abs(z_score)))

# Make decision
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

# Visualize the decision region
x_min = -3
x_max = 3
x = np.linspace(x_min, x_max, 1000)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, color='blue', label='Standard Normal Distribution')

# Shade the rejection regions
x_critical_left = np.linspace(x_min, -critical_value, 100)
y_critical_left = norm.pdf(x_critical_left, 0, 1)
plt.fill_between(x_critical_left, y_critical_left, color='red', alpha=0.5, label='Rejection Region')

x_critical_right = np.linspace(critical_value, x_max, 100)
y_critical_right = norm.pdf(x_critical_right, 0, 1)
plt.fill_between(x_critical_right, y_critical_right, color='red', alpha=0.5)

# Plot the Z-score
plt.axvline(z_score, color='green', linestyle='dashed', linewidth=1, label=f'Z-score = {z_score:.2f}')

plt.title('Two-Tailed Z-test')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ques 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

# Solution 5.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0, mu1, sigma, alpha):
    """
    Calculates and visualizes Type 1 and Type 2 errors in hypothesis testing.

    Args:
        mu0: Mean of the null hypothesis distribution.
        mu1: Mean of the alternative hypothesis distribution.
        sigma: Standard deviation of both distributions.
        alpha: Significance level (Type 1 error rate).
    """

    # Calculate the critical value (z-score)
    z_critical = norm.ppf(1 - alpha)  # Use inverse CDF to get the z-score

    # Calculate the critical value in terms of the sample mean (x_bar)
    x_bar_critical = mu0 + z_critical * sigma

    # Calculate Type 1 error (false positive)
    type1_error = alpha

    # Calculate Type 2 error (false negative)
    type2_error = norm.cdf( (x_bar_critical - mu1) / sigma )

    # Plot the distributions
    x = np.linspace(mu0 - 5 * sigma, mu0 + 5 * sigma, 100)  # Range for plotting
    plt.plot(x, norm.pdf(x, mu0, sigma), label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, norm.pdf(x, mu1, sigma), label='Alternative Hypothesis (H1)', color='red', linestyle='dashed')

    # Mark the critical value
    plt.axvline(x_bar_critical, color='green', linestyle='dotted', label=f'Critical Value: x_bar = {x_bar_critical:.2f}')

    # Shade the Type 1 error region
    plt.fill_between(x, norm.pdf(x, mu0, sigma), where=(x > x_bar_critical), color='orange', alpha=0.3, label=f'Type I Error (False Positive) = {type1_error:.2%}')

    # Shade the Type 2 error region
    plt.fill_between(x, norm.pdf(x, mu1, sigma), where=(x < x_bar_critical), color='pink', alpha=0.3, label=f'Type II Error (False Negative) = {type2_error:.2%}')

    # Add labels and title
    plt.xlabel('Sample Mean (x_bar)')
    plt.ylabel('Probability Density')
    plt.title('Type I and Type II Error Visualization')

    # Add legend
    plt.legend(loc='upper right')

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage:
if __name__ == "__main__":
    mu0 = 10  # Mean of the null hypothesis distribution
    mu1 = 11  # Mean of the alternative hypothesis distribution
    sigma = 2  # Standard deviation
    alpha = 0.05  # Significance level (Type 1 error rate)

    visualize_errors(mu0, mu1, sigma, alpha)
```

```python
# Ques 6. Write a Python program to perform an independent T-test and interpret the results.

# Solution 6.

import numpy as np
from scipy import stats

# Sample data for two independent groups
group1 = np.array([25, 30, 28, 35, 22, 29, 31, 27, 33, 26])
group2 = np.array([38, 42, 40, 35, 45, 39, 41, 37, 43, 44])

# Perform independent T-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Set significance level (alpha)
alpha = 0.05

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("There is a significant difference between the means of the two groups.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is no significant difference between the means of the two groups.")
```

```python
# Ques 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

# Solution 9.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data: A list or NumPy array of sample data.
        confidence: The desired confidence level (e.g., 0.95 for 95% confidence).

    Returns:
        A tuple containing the lower and upper bounds of the confidence interval.
    """
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1) # ddof=1 for sample standard deviation

    # Calculate the critical t-value
    t_critical = t.ppf((1 + confidence) / 2, n - 1)

    # Calculate the margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example Usage:
data = [23, 25, 28, 30, 32, 35, 37, 39]
lower, upper = confidence_interval(data, confidence=0.95)
print(f"The 95% confidence interval for the mean is: ({lower:.2f}, {upper:.2f})")
```

```python
# Ques 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

# Solution 10.


import scipy.stats as st

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level.

    Args:
        data (list or numpy.ndarray): Sample data.
        confidence (float, optional): Confidence level (e.g., 0.95 for 95% confidence).
                                     Defaults to 0.95.

    Returns:
        float: Margin of error.
    """
    n = len(data)
    mean = np.mean(data)
    std_err = st.sem(data)

    if std_err == 0:
      return 0.0

    interval = st.t.interval(confidence, n - 1, loc=mean, scale=std_err)
    margin_of_error = (interval[1] - interval[0]) / 2
    return margin_of_error

if __name__ == '__main__':
    data = [85, 90, 88, 92, 87, 93, 89, 91, 86, 94]
    confidence_level = 0.95
    margin_error = calculate_margin_of_error(data, confidence_level)
    print(f"Margin of Error at {confidence_level*100}% Confidence Level: {margin_error:.2f}")

    data2 = [1,2,3,4,5]
    confidence_level2 = 0.99
    margin_error2 = calculate_margin_of_error(data2, confidence_level2)
    print(f"Margin of Error at {confidence_level2*100}% Confidence Level: {margin_error2:.2f}")
```

```python
# Ques 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

# Solution 11.

def bayes_theorem(prior, likelihood, false_positive_rate, evidence):
    # Calculate the marginal likelihood (P(B))
    marginal_likelihood = (likelihood * prior) + (false_positive_rate * (1 - prior))

    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior

# Given probabilities
prior = 0.01  # P(Disease)
likelihood = 0.95  # P(Pos|Disease)
false_positive_rate = 0.05  # P(Pos|No Disease)

# Calculate posterior probability
posterior = bayes_theorem(prior, likelihood, false_positive_rate, evidence=None)
print(f"The probability of having the disease given a positive test result is: {posterior:.4f}")

```

```python
# Ques 12. #12 Perform a Chi-square test for independence between two categorial variables in python

# Solution 12.

import numpy as np
from scipy.stats import chi2_contingency

# Observed data
observed = np.array([[20, 15, 10],  # Male
                     [25, 20, 30]]) # Female

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

# Display results
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: Variables are dependent.")
else:
    print("Fail to reject the null hypothesis: Variables are independent.")
```

```python
# Ques 13. #13 Write a python program to calculate the expected frequencies for a Chi-square test based on observed data

# Solution 13.

import numpy as np

# Observed data
observed = np.array([[20, 15, 10],  # Male
                     [25, 20, 30]]) # Female

# Calculate row totals and column totals
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

# Initialize expected frequency table
expected = np.outer(row_totals, col_totals) / grand_total

# Display expected frequencies
print("Expected Frequencies:")
print(expected)
```
